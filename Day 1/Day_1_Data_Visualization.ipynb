```python
# We will see the trend/pattern
plt.plot(Salary[0])
plt.xlabel("Season")
plt.ylabel("Salary")
plt.show()
```

```python
# Providing the title
plt.plot(Salary[0], ls = '-.', color ='red')
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.show()
```

```python
# Ticklabel is plain
plt.plot(Salary[0], ls = '-.', color ='red')
plt.ticklabel_format(style='plain')
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.show()
```

```python
# Default ticklabel is sci
plt.plot(Salary[0], ls = '-.', color ='red')
plt.ticklabel_format(style='sci')
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.show()
```

```python
# xticks - plt.xticks(old,new)
plt.plot(Salary[0], ls = '-.', color ='red')
plt.ticklabel_format(style='plain')
plt.xticks(list(range(10)), Seasons)
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.show()
```

```python
# xticks - plt.xticks(old,new)
plt.plot(Salary[0], ls = '-.', color ='red')
plt.ticklabel_format(style='plain')
plt.xticks(list(range(10)), Seasons, rotation = 90)
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.show()
```

```python
# xticks - plt.xticks(old,new)
plt.plot(Salary[0], ls = '-.', color ='red')
plt.ticklabel_format(style='plain')
plt.xticks(list(range(10)), Seasons, rotation = 20)
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.show()
```

```python
# Marker style - . , o .....
plt.plot(Salary[0], ls = '-.', color ='red', marker = 'o')
plt.ticklabel_format(style='plain')
plt.xticks(list(range(10)), Seasons, rotation = 20)
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.show()
```

```python
# Marker style - . , o 
plt.plot(Salary[0], ls = '-.', color ='red', marker = '*')
plt.ticklabel_format(style='plain')
plt.xticks(list(range(10)), Seasons, rotation = 20)
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.show()
```

```python
# Marker style - . , o 
plt.plot(Salary[0], ls = '-.', color ='red', marker = '^')
plt.ticklabel_format(style='plain')
plt.xticks(list(range(10)), Seasons, rotation = 20)
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.show()
```

```python
# Including the player 2
plt.plot(Salary[0], ls = '-.', color ='red', marker = '^')
plt.plot(Salary[1], ls = '--', color ='green', marker = 'P')
plt.ticklabel_format(style='plain')
plt.xticks(list(range(10)), Seasons, rotation = 20)
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.show()
```

```python
# labels
plt.plot(Salary[0], ls = '-.', color ='red', marker = '^', label = Players[0])
plt.plot(Salary[1], ls = '--', color ='green', marker = 'P', label = Players[1])
plt.ticklabel_format(style='plain')
plt.xticks(list(range(10)), Seasons, rotation = 20)
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.legend()
plt.show()
```

```python
# bbox_to_anchor=(1,1)
plt.plot(Salary[0], ls = '-.', color ='red', marker = '^', label = Players[0])
plt.plot(Salary[1], ls = '--', color ='green', marker = 'P', label = Players[1])
plt.ticklabel_format(style='plain')
plt.xticks(list(range(10)), Seasons, rotation = 20)
plt.xlabel("Season")
plt.ylabel("Salary")
plt.title("Players Analysis")
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```
