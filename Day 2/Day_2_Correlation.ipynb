# Correlation Calculation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
DF = pd.read_csv(r"DemographicData.csv")
DF
```

```python
DF.columns = (['CountryName', 'CountryCode', 'BirthRate', 'InternetAccess',
       'IncomeGroup'])
DF
```

# Correlation Calculation between X & y

```python
from scipy.stats.stats import pearsonr
```

## pearsonr = 1.0   X & y are 100% correlated   
## p-value = 0.0    X & y are 0% uncorrelated

```python
X = [1,2,3,4,5]
y = [2,4,6,8,10] 

pearsonr(X, y)
```

```python
plt.plot(X, y, ls = '-.', color ='red')
plt.xlabel("X")
plt.ylabel("y")
plt.title("X vs y")
plt.show()
```

## pearsonr = 0.008   X & y are 0% correlated   
## p-value = 0.98    X & y are 98% uncorrelated

```python
X = [0,7,11,1,-5]
y = [-2,2000,-1000,-11,0] 

pearsonr(X, y)
```

```python
plt.plot(X, y, ls = '-.', color ='red')
plt.xlabel("X")
plt.ylabel("y")
plt.title("X vs y")
plt.show()
```

## Correlation Calculation between X as InternetAccess &  y as BirthRate

## pearsonr = -0.81   X & y are 81%  negatively correlated   
## p-value = 0.0    X & y are 0% uncorrelated

```python
pearsonr(DF.InternetAccess, DF.BirthRate)
```

```python
print(f"P value: {pearsonr(DF.InternetAccess, DF.BirthRate)[0]}")
```

## lmplot() for correlation

Linear model plot

```python
sns.lmplot("InternetAccess","BirthRate",data=DF)
plt.show()
```

```python
sns.lmplot("InternetAccess","BirthRate",data=DF, fit_reg=True,
          hue="IncomeGroup")

plt.show()
```

## hue is used for grouping the 

## fit_reg is used to draw the trend line

```python
sns.lmplot("InternetAccess","BirthRate",data=DF, fit_reg=False,
          hue="IncomeGroup")

plt.show()
```

Most of the datapoints have high income group and low birth rate.

Most of the datapoints which have low income group have high birth rate.
