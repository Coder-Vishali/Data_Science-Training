# Type casting columns 

```python
empdf.info()
```

```python
def convertInt(x):
    if x == "Null":
        return 0
    else:
        return int(x)
```

```python
def convertFloat(x):
    if x == "Null":
        return 0.0
    else:
        return float(x)
```

```python
empdf.department_id = empdf.department_id.apply(convertInt)
empdf.manager_id = empdf.manager_id.apply(convertInt)
empdf.commission_pct = empdf.commission_pct.apply(convertFloat)
```

```python
empdf.info()
```

```python
empdf.describe() # It picks up all the numeric columns now
```
